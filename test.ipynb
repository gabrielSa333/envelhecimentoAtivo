```python
import pandas as pd

df = pd.read_csv("data/processed/pns_2019_core_corrected.csv")    
df.head(5)
```

```python
# quero ver todas as colunas sem supressao visual
pd.set_option('display.max_columns', None)
df.head(5)
```

```python
# Análise da variável dependencia_SUS
print("Distribuição de dependencia_SUS:")
print(df['dependencia_SUS'].value_counts())
print("\nProporção:")
print(df['dependencia_SUS'].value_counts(normalize=True))
print("\nValores únicos:")
print(df['dependencia_SUS'].unique())
print("\nDescrição estatística:")
print(df['dependencia_SUS'].describe())
```

```python
# Verificar a lógica: crosstab entre possui_plano_saude e atendimento_sus
print("Crosstab possui_plano_saude vs atendimento_sus:")
print(pd.crosstab(df['possui_plano_saude'], df['atendimento_sus'], margins=True))
print("\nValores únicos de possui_plano_saude:", df['possui_plano_saude'].unique())
print("Valores únicos de atendimento_sus:", df['atendimento_sus'].unique())
```

```python
# Explicação das últimas colunas: autoav_z, multimorb_z, functional_z, health_score_raw, health_score
print("=== EXPLICAÇÃO DAS ÚLTIMAS COLUNAS ===")
print("\n1. autoav_z: Z-score da autoavaliação de saúde (autoavaliacao_saude)")
print("   - Padroniza a autoavaliação para média 0 e desvio padrão 1")
print("   - Valores negativos = saúde melhor que a média, positivos = pior")
print("\n2. multimorb_z: Z-score da contagem de multimorbidades (multimorbidity_count)")
print("   - Padroniza a contagem de doenças para média 0 e desvio padrão 1")
print("   - Valores negativos = menos doenças que a média, positivos = mais")
print("\n3. functional_z: Z-score do escore funcional (functional_score)")
print("   - Padroniza o escore de capacidade funcional para média 0 e desvio padrão 1")
print("   - Valores negativos = melhor funcionalidade que a média, positivos = pior")
print("\n4. health_score_raw: Escore bruto de saúde composto")
print("   - Fórmula: (-0.5 * autoav_z) + (-0.7 * multimorb_z) + (1.2 * functional_z)")
print("   - Combina os três indicadores com pesos específicos")
print("   - Valores mais altos = pior saúde geral")
print("\n5. health_score: Escore de saúde normalizado (0-1)")
print("   - health_score_raw escalado para o intervalo [0,1]")
print("   - 0 = melhor saúde possível, 1 = pior saúde possível")
print("\n=== ESTATÍSTICAS DESCRITIVAS ===")
print("\nautoav_z:")
print(df['autoav_z'].describe())
print("\nmultimorb_z:")
print(df['multimorb_z'].describe())
print("\nfunctional_z:")
print(df['functional_z'].describe())
print("\nhealth_score_raw:")
print(df['health_score_raw'].describe())
print("\nhealth_score:")
print(df['health_score'].describe())
```
